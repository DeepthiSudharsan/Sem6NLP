### 3. Choose a corpus with more than 10 sentences. Write a python code to create a co- occurrence matrix (choose window length as 1). Apply SVD. Perform dimensionality reduction operation. Note down the observations about any relations between words.

```python
# splitting each sentence as list of words
word_line = []
for l in lines:
    word_line.append(l.split())
word_line
```

```python
# since window size is 1 if a sentence has the particular word, if there is a word in front of it in the sentence
# we add 1 to its column in that word's row and similarly if there is a word after the particular word then we do
# the same and we get the cooccurence matrix
cooccur = np.zeros((len(word_corpus),len(word_corpus)),dtype = int)
for word in word_corpus:
    for sent in word_line:
        if word in sent:
            if sent.index(word) - 1 >= 0:
                cooccur[list(word_corpus).index(word)][list(word_corpus).index(sent[sent.index(word) - 1])] += 1
            if sent.index(word) + 1 < len(list(sent)):
                cooccur[list(word_corpus).index(word)][list(word_corpus).index(sent[sent.index(word) + 1])] += 1
# displaying coocurence matrix as a dataframe
cooccur_matrix = pd.DataFrame(cooccur, columns = word_corpus)
cooccur_matrix.index = word_corpus
cooccur_matrix
```

```python
# performing svd in cooccurence matrix
U,S,V = svd(cooccur)
S # Singular Values
```

```python
S_rounded = np.floor(S) # rounding off the singular valuesto the lowest closest value
k = np.count_nonzero(S_rounded) # approximate number of singular values that are large and not close to 0
Uk = U[:,:k] # eigen vectors
Uk
```

```python
print("Original feature dimensions",U.shape)
print("Reduced feature dimensions",Uk.shape)
```

```python
from sklearn.decomposition import TruncatedSVD
#using built in and doing the dimensionality reduction
svd = TruncatedSVD(n_components = k)
cooccur_truncated = svd.fit_transform(cooccur)
cooccur_truncated
```

#### We notice that unlike the cooccurence matrix that had binary values, here we have real values (both negative and positive), this shows that during the process of dimensionality reduction, smoothening has happened and we can better observe even the smallest of correlations between different words.
